```python
# Handle necessary imports
import pandas as pd
import numpy as np
```

```python
# Load dataset
train = pd.read_csv('train.csv')
```

# 1. Constant bidding

```python
def constant_bidding(constant):
    
    impressions = 0.0
    clicks = 0.0
    costs = 0.0
    budgets = 25000000
    
    for click, pay_price in train[['click','payprice']].values:
        
            if constant > pay_price:
                impressions += 1.0
                clicks += click
                costs += pay_price
                
            if costs >= budgets:
                break
                
    return impressions, clicks, costs

const_bid = pd.DataFrame()
const_bid['constants'] = np.arange(40,400,20)

impression = []
click = []
cost_sum = []

for constt in const_bid['constants']:
    
    [imp_const, click_const, cost_const] = constant_bidding(constt)
    impression.append(imp_const)
    click.append(click_const)
    cost_sum.append(cost_const)
    
const_bid['imp_const_win'] = impression
const_bid['click_const_win'] = click
const_bid['CTR'] = (const_bid.click_const_win/const_bid.imp_const_win * 100).round(7).astype(str)
const_bid['cost_const_sum'] = cost_sum
const_bid['CPM'] = (const_bid.cost_const_sum/const_bid.imp_const_win * 1000).round(2).astype(str)
const_bid['CPC'] = (const_bid.cost_const_sum/const_bid.click_const_win).round(2).astype(str)
```

# 2. Random bidding

```python
from random import randrange

def random_bidding(upper_bound):
    impressions = 0.0
    clicks = 0
    costs = 0.0
    budgets = 25000000
    rand_no = randrange(upper_bound)
    
    for click,pay_price in train[['click','payprice']].values:  
        if rand_no > pay_price:
            impressions += 1
            clicks += click
            costs += pay_price
        if costs >= budgets:
                break
    return rand_no,impressions, clicks, costs
            
randm_bid = pd.DataFrame()
randm_bid['upper_bound'] = np.arange(200,390,10)

rands = []
im = []
clks = []
ct = []
for upper in randm_bid['upper_bound']:
    [rans,imps, clicks, cost] = random_bidding(upper)
    rands.append(rans)
    im.append(imps)
    clks.append(clicks)
    ct.append(cost)
randm_bid['random_no'] = rands
randm_bid['impression_randm_win'] = im
randm_bid['click_randm_win'] = clks
randm_bid['CTR'] = (randm_bid.click_randm_win/randm_bid.impression_randm_win * 100).round(7).astype(str)
randm_bid['cost_randm_sum'] = ct
randm_bid['CPM'] = (randm_bid.cost_randm_sum/randm_bid.impression_randm_win * 1000).round(2).astype(str)
randm_bid['CPC'] = (randm_bid.cost_randm_sum/randm_bid.click_randm_win).round(2).astype(str)
```
